### 학습 목표
- K-최근접 이웃 (K Nearest Neighbors), KNN 모델로 와인 데이터셋을 분석해 와인 등급을 예측하고 작동 원리를 이해하자.
- 종속변수가 범주형이며, 개수가 3개 이상인 다중분류(Multicalssification)

### 학습 순서
- 문제 정의
- 라이브러리 및 데이터 불러오기
- 데이터 확인하기
- 목푯값에서 고윳값 확인하기
- 전처리 : 결측치 처리하기
- 스케일링
- 모델링 및 예측/평가하기
- 하이퍼파라미터 튜닝하기
- 이해하기 : K-최근접 이웃

### K-최근접 이웃 소개
- 거리 기반 모델
- 지금까지 다룬 알고리즘들과 달리 선형 관계를 전제로 하지 않는다.
- 즉 각 데이터 간의 거리를 활용해서 새로운 데이터를 예측하는 모델
  - 데이터 간의 거리를 활용해서 새로운 데이터가 어디에 속하는지 구분하는 알고리즘
  - 이때 가까이에 있는 데이터를 고려하여 예측값이 결정된다
  - K Nearest Neighbors라는 이름은 이를 잘 반영하는데, K개의 가장 가까운 이웃 데이터에 의해 예측된다는 의미
- 다중분류에서 가장 간편히 적용할 수 있는 알고리즘

### 장점
- 수식에 대한 설명이 필요 없을 만큼 직관적이고 간단
- 선형 모델과 다르게 별도의 가정이 있다
  - 예를 들어, 선형 회귀는 독립변수와 종속 변수의 선형 관계를 가정하고 있기 때문에, 이 가정이 들어맞지 않는 데이터에 취약하나, KNN은 이러한 가정이 없어 더 자유롭다

### 단점
- 데이터가 커질수록 느려질 수 있다
- 아웃라이어에 취약하다
  - 아웃라이어 : 평균치에서 크게 벗어나는 데이터, 이상치라고도 한다.
    - 예를 들어, 한국 축구 선수들의 연봉 데이터가 있을 때, 평균 연봉은 약 2억이고 스타 플레이어는 150억 이상인 경우 스타 플레이어의 연봉을 아웃라이어라고 한다

### 유용한 곳
- 주로 분류에서 사용되며, 로지스틱 회귀로 해결할 수 없는 3개 이상의 목표변수들도 분류할 수 있다.
- 작은 데이터셋에 적합

### 문제 정의
- 데이터셋 소개 : 와인에 대한 데이터, 총 3가지 목푯값으로 이루어진 범주형 변수이기 때문에 다중분류(Multicalssificaion)문제에 해당한다. 알코올, 말산, 마그네슘, 색조 등이 독립변수로, 와인 등급인 class를 종속변수로 사용한다.

## 6.6 스케일링

- 스케일링은 데이터 스케일(Scale)을 맞추는 작업
- describe를 호출해 데이터를 보았을 때, 각 컬럼들마다 값들의 범위가 다양하다
  - K-최근접 이웃은 거리 기반의 알고리즘이기 때문에, 이러한 스케일 문제가 좋지 않은 결과 초래할 수 있다.
  - 즉 alcohol에서의 1과 magnesium에서의 1은 완전히 다른 영향을 미치기 때문에, KNN알고리즘이 왜곡된 예측을 할 수 있다.
- 스케일링은 이러한 문제를 해결할 목적으로 인위적으로 각 컬럼이 비슷한 범위를 가지도록 만드는 작업

### 종류
- 표준화 스케일링 : 평균이 0이 되고, 표준편차가 1이 되도록 데이터를 고르게 분포시키는 데 사용
- 로버스트 스케일링 : 데이터에 아웃라이어가 존재하고, 그 영향력을 그대로 유지하고 싶을 때 사용
- 최소-최대 스케일링 : 데이터 분포의 특성을 최대한 그대로 유지하고 싶을 때 사용
- 정규화 스케일링 : 행 기준의 스케일링이 필요할 때 사용하나, 실제로 거의 사용하지 않음

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
```

```python
st_scaler = StandardScaler() # 스케일러 지정
```

```python
# fit 함수로 data를 학습
# => 스케일링에 필요한 정보 (평균, 표준편차)가 학습된다.
st_scaler.fit(data)
```

```python
# transform 으로 연산
# => 이 단계에서는 학습에서 얻은 정보로 계산하게 된다.
st_scaled = st_scaler.transform(data)

# transform() : 스케일러가 fit을 통해 학습한 정보를 통해 데이터를 변환, 즉 스케일링하는 함수
```

```python
# 넘파이 배열
# => 계산을 더 효율적으로 할 목적으로 컬럼명 없이 데이터를 한 줄로 쭉 이어서 표현
# => 인덱스와 컬럼명 없이 한 줄로 표현 => 컴퓨터가 계산하기 쉽다.
st_scaled
```

```python
# 사람이 분석하기에는 넘파이보단 판다스가 편함, 데이터를 판다스로 변경
pd.DataFrame(st_scaled)

# 배열에는 인덱스와 컬럼명이 없어서 데이터프레임으로 변환할 때, 각각의 숫자가 자동으로 적용된다.
```

```python
# 컬럼명을 지정하여 데이터 프레임으로 변환
st_scaled = pd.DataFrame(st_scaled, columns = data.columns)
```

```python
round(st_scaled.describe(), 2)

# 모든 컬럼에서 평균 0, 표준편차 1
# min과 max값은 컬럼마다 각기 다르지만, 처음보다는 훨씬 줄었다.
# => 데이터가 더 동등한 수준에서 연산될 수 있다.
```

### 이해하기 : 표준화 스케일링
- 데이터를 표준화된 정규분포 (Standard Normal Distribution) 로 만들어주는 방법
- 각 변수를 기준으로 하여, 각 데이터에 해당 변수의 평균만큼을 빼주고 이를 표준편차로 나눈다
  - 이 수식을 거치면 모든 컬럼이 표준정규분포의 형태를 따른다
  - (xi - mean(x)) / sd(x)
- fit() 함수로 학습을 시켜주는 과정에서 각 컬럼의 평균과 표준편차가 st_scaler에 기억되고, transform()을 적용하면 그 값들을 이용하여 위의 수식으로 연산하는 것
- 데이터를 표준화한다면 평균은 0이고 분산은 1인 형태로 데이터가 변경된다.

```python
rb_scaler = RobustScaler()
```

```python
rb_scaled = rb_scaler.fit_transform(data)
rb_scaled = pd.DataFrame(rb_scaled, columns = data.columns)
```

```python
round(rb_scaled.describe(), 2)
```

### 이해하기 : 로버스트 스케일링
- 로버스트 스케일링은 평균과 표준편차 대신 사분위값을 이용하여 계산
- 평균값이 0에 가깞지만 완전 0은 아니고, 분산 또는 표준화 스케일링과는 다르게 1로 고정되지 않았다.
- (xi - Q2(x)) / (Q3(x) - Q1(x))
  - Q2(x) 는 50% 지점인 중윗값

```python
mm_scaler = MinMaxScaler()
mm_scaled = mm_scaler.fit_transform(data)
mm_scaled = pd.DataFrame(mm_scaled, columns = data.columns)
round(mm_scaled.describe(), 2)
```

### 이해하기 : 최소-최대 스케일링
- 모든 컬럼에서 최댓값이 1, 최솟값이 0 인 형태로 변환
- (xi - min(x)) / (max(x) - min(x))

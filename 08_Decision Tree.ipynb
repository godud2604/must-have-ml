```python
# 그룹별 class의 평균값을 계산하여 저장
# 국가 이름이 키값이 되어야 하는데, 현재 country_group에는 국가 이름은 변수가 아닌 인덱스에 자리하고 있다. reset_index()를 써서 변수로 뺄 수 있다.
country_group = data.groupby('native-country').mean()['class']
country_group
```

```python
# reset_index()를 써서 인덱스를 변수로 빼냄
country_group = country_group.reset_index()
```

```python
country_group
```

```python
# merge를 이용하여 두 데이터를 합침
# data와 country_group을 native_country 기준으로 합침
data = data.merge(country_group, on = 'native-country', how='left')
```

```python
data
```

```python
data.drop('native-country', axis=1, inplace=True)
# rename: 컬럼 이름을 바꾸는데 사용
# => 변경할 대상(columns 혹은 index)을 지정해주고, 구체적인 변경 사항을 딕셔너리 형태로 매핑
data = data.rename(columns= {'class_x': 'class', 'class_y': 'native-country'})
```

## 8.4 전처리 : 결측치 처리 및 더미 변수 변환

```python
# 결측치를 처리
data.isna().mean() # 1. 결측치 비율 확인

# native-country : 각 국가별 class의 평균값으로 대체한 상태
# => 원래는 범주형 변수이기 때문에, 평균치로 채우는 것이 불가능하지만 지금은 숫자로 바뀌었고 그 숫자가 마침 class의 평균값이기 때문에, mean()이나 median()으로 결측치를 채우는 것도 가능하다.
# => 또는 완전 별개의 숫자를 지정하여 채우는 방법도 있다. 보통 결측치를 숫자로 채워줄 때 -9와 -99같은 임의의 숫자를 사용한다.
# => 임의의 숫자는, 트리 기반 모델에서는 큰 문제가 없지만, 선형 모델에서는 데이터의 왜곡을 불러오니 주의
```

```python
# 결측치를 -99로 대체
data['native-country'] = data['native-country'].fillna(-99)
```

```python
data['workclass'].value_counts()
```

```python
# 결측치를 private로 대체
data['workclass'] = data['workclass'].fillna('Private')
```

```python
data['occupation'].value_counts()
```

```python
# 결측치를 Unknown으로 대체
data['occupation'] = data['occupation'].fillna('Unknown')
```

```python
# 모든 결측치 해결, 범주형 데이터를 더미 변수로 변환 ?
data = pd.get_dummies(data, drop_first=True)
```

## 8.5 모델링 및 평가하기

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data.drop('class', axis=1), data['class'], test_size = 0.4, random_state = 100)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

### 8.6 이해하기
- 결정 트리는 특정 변수에 대한 특정 기준값으로 데이터를 계속 분류해가면서 유사한 그룹으로 묶어내어 예측값을 만드는 알고리즘이다.
- 좋은 모델을 만들려면 첫 번째 분류에 사용할 변수 선정과 기준점 정하기가 중요, 그리고 두 번째 분류 과정에서도 각 상황에서 최적의 변수와 기준점을 채택해야 한다.

### 8.6.1 분류 결정 트리
- 지니 인덱스와(Gini index) 교체 엔트로피(Cross Entropy)
  - 노드의 순도를 평가하는 방법, 노드의 순도가 높을수록 지니 및 엔트로피 값은 낮아진다.
- 지니 인덱스 : 각 노드에 대해서 계산된다.
  - 결정 트리에서는 지니 인덱스가 가장 낮은 값이 나오는 특정 변수의 특정 값을 기준으로 노드를 분류한다
- 교차 엔트로피
  - 마찬가지로 순도가 높을수록 낮은 값을 보이나, 최댓값은 1까지 나올 수 있다는 차이점이 있다.

### 8.6.2 회귀 결정 트리
- 회귀는 MSE를 평가 기준으로 이용한다.
  - MSE : 실젯값과 예측값의 차이에 대한 계산 ( y: 실젯값, y의 평균: 예측값 )
- 결정트리 회귀는 가장 낮은 MSE값이 나오도록 노드를 분류해나간다.

### 8.7 오버피팅 문제
- 훈련셋을 사용하여 지나치게 깊은 트리 모델을 만들면, 시험셋에 대한 예측은 오히려 떨어지는 경향이 있다.
  - 훈련셋을 지나치게 잘 예측하는 모델의 정확도를 일부러 낮춰 오버피팅을 예방할 필요가 있다.

- 언더피팅 : 과소적합이라고도 하며, 모델이 충분히 학습되지 않아 훈련셋에 대해서도 좋은 예측력을 내지 못하지는 상황

- 편향-분산 트레이드오프(bias-variance tradeoff) : 훈련셋을 지나치게 잘 예측하는 모델의 정확도를 일부러 낮춰 오버피팅을 예방할 필요가 있다.
- 편향
  - 독립변수와 종속변수를 모델링한 알고리즘이 적절치 못하거나 중요한 부분을 놓쳐서 제대로 된 예측을 하지 못할 때 높아진다.
  - 예를 들어, 앞서 봤던 언더피팅 예시가 편향이 높은 예시이다.
- 분산
  - 훈련셋에 있는 데이터의 노이즈에 의해 발생하는 오차
  - 예를 들어 앞서 살펴본 오버피팅은 알고리즘이 훈련셋을 지나치게 정확하게 따라가면서 모든 노이즈를 모델에 포함해 높은 분산이 발생할 수 있다.
- 결론적으로는 높은 편향은 언더피팅 문제를, 높은 분산은 오버피팅문제를 불러올 수있다.
  - 각 오차들의 합이 전체 모델의 에러가 되기 때문에 이 둘의 합이 가장 낮은 중간 지점이 우리가 찾아야 할 최적점이다.

## 8.8 매개변수 튜닝

```python
# 결정 트리에서는 트리 깊이가 깊어질수록, 즉 수없이 많은 노드를 분류하여 모델을 만들수록 오버피팅이 발생할 가능성이 높다
# => 이 문제를 해결할 목적으로, 결정 트리에서는 트리의 깊이를 제한하는 매개변수를 제공한다.
# => 기본값이 None이므로, 매개변수를 지정해주지 않으면 최대한 깊은 수준으로 트리를 만듦

model = DecisionTreeClassifier() # 모델 객체 생성
model.fit(X_train, y_train) # 훈련셋 학습
train_pred = model.predict(X_train) # 훈련셋 예측
test_pred = model.predict(X_test) # 시험셋 예측
print('Train score:', accuracy_score(y_train, train_pred), 'Test score:', accuracy_score(y_test, test_pred)) # 훈련셋, 시험셋의 정확도 평가
```

```python
model = DecisionTreeClassifier(max_depth=5)
model.fit(X_train, y_train)
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)
print('Train score:', accuracy_score(y_train, train_pred), 'Test score:', accuracy_score(y_test, test_pred))
```

```python
model = DecisionTreeClassifier(max_depth=7)
model.fit(X_train, y_train)
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)
print('Train score:', accuracy_score(y_train, train_pred), 'Test score:', accuracy_score(y_test, test_pred))
```
